```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
measure = [24.23,25.53,25.41,24.14,29.62,28.25,25.81,24.39,40.26,32.95,91.36,25.99,39.42,26.71,35.00]
company = ['Allied Signal','Bankers Trust','General Mills','ITT Industries','J.P.Morgan & Co.','Lehman Brothers','Marriott','MCI','Merrill Lynch','Microsoft','Morgan Stanley','Sun Microsystems','Travelers','US Airways','Warner-Lambert']
```

```python
measure_x =pd.Series(measure)
```

```python
plt.figure(figsize=(10,9))
ax = sns.barplot(y=measure,x=company)
plt.xticks(rotation=90)
for i in ax.containers:
    ax.bar_label(i)
plt.show()
```

```python
sns.boxplot(measure_x)
```

```python
Q1 = np.quantile(measure_x,0.25)
Q3 = np.quantile(measure_x,0.75)
med = np.median(measure_x)
IQR = Q3 - Q1
upper_bound = Q3 + (1.5*IQR)
lower_bound = Q1 - (1.5*IQR)
Outliers = measure_x[(measure_x <= lower_bound) | (measure_x >= upper_bound)]
print('First Quantile=', Q1, 'Second Quantile=', med, 'Third Quantile=', Q3,
      'Inter-Quartile Range=', IQR, 'Upper Whisker=', upper_bound, 'Lower Whisker=', lower_bound)
print('The outlier in the boxplot:',Outliers)
Outliers = measure_x[(measure_x >= upper_bound)]
print('The outlier in the boxplot:',Outliers)
print(measure_x.describe())
print("Variance =",measure_x.var())
```
